### **🔶 Task 1/3: DNA to RNA Transcription**
Use the requests `get` function to retrieve a fasta file from uniprot then save it into a fasta file called hemoglobin.fasta



````python
import requests

# Defining url and pdb_id
uniprot_id = "P68871"
url = f"https://rest.uniprot.org/uniprotkb/{uniprot_id}.fasta"
response = requests.get(url) #request the data

# Write into a file called "hemoglobin.fasta the" contents of the request"
with open("hemoglobin.fasta", "w") as f:
  #write the contents of the response into the file f (Use the .write method to write in the data)
  #(The contents of the response are found in the .text part of the object)
    f.write(response.text)

print(response.text)
print("Saved as hemoglobin.fasta")
````

<details>
<summary><font color="orange">Click here for useful information about amino acids</font></summary>

1. What does the composition reveal about the properties of the protein?
The amino acid composition reveals key features of a protein's structure and function:

Hydrophobic vs. Hydrophilic Residues: A higher concentration of hydrophobic amino acids suggests the protein may have hydrophobic regions, which are important for membrane association or folding into a compact structure. A predominance of hydrophilic residues indicates the protein is likely soluble in aqueous environments.

Charge Distribution: Charged amino acids (e.g., lysine, glutamate) often suggest the protein may interact with charged molecules or ions, aiding in stability or function.

Functional Significance: Certain amino acids, like histidine in enzyme active sites or cysteine in disulfide bonds, highlight the protein's specific biological role.

2. How might amino acid frequencies influence protein folding and function?
Amino acid frequencies influence protein folding and function in the following ways:

Hydrophobic Core Formation: Hydrophobic amino acids tend to cluster inside, stabilizing the protein’s structure by minimizing the system's energy.

Hydrophilic Surface Exposure: Hydrophilic amino acids are more likely to be exposed on the surface, aiding interactions with water or other molecules.

Secondary Structure Formation: Amino acids like proline and glycine can influence secondary structure formation, such as alpha-helices and beta-sheets, affecting protein flexibility and function.

Functional Sites: The presence of certain residues at specific positions, like serine or histidine in enzyme active sites, is crucial for catalytic activity.

3. What is the role of specific amino acids in stabilizing protein structures?
Specific amino acids stabilize protein structures through various interactions:

Disulfide Bonds (Cysteine): These covalent bonds stabilize the protein’s tertiary and quaternary structure, especially in extracellular proteins.
Salt Bridges (Lysine, Glutamate, Arginine, Aspartate): Charged amino acids form salt bridges that stabilize the folded structure.

Hydrogen Bonds (Serine, Threonine, Asparagine): Polar residues form hydrogen bonds, which maintain secondary structure and enable protein-protein interactions.
Aromatic Stacking (Phenylalanine, Tyrosine, Tryptophan): Aromatic residues stabilize protein structure by reducing steric strain and promoting compact folding.

These interactions ensure the protein's functional stability, and disruptions can lead to misfolding or denaturation, potentially resulting in diseases.
</details>

````python
from Bio import SeqIO
from collections import Counter
    # parse fasta file
filename = "hemoglobin.fasta"  # Replace with your actual file name

for record in SeqIO.parse(filename, "fasta"):
    protein_id = record.id
    sequence = str(record.seq)

    # Count amino acid frequencies
    aa_count = Counter(sequence)
    total_residues = len(sequence)

    print(f"Protein ID: {protein_id}")
    print(f"Total Length: {total_residues}")
    print("Amino Acid Frequencies:")
    for aa, count in aa_count.items():
        frequency = (count / total_residues) * 100
        print(f"{aa}: {count} ({frequency:.2f}%)")
````

###**🔶Task 3/3. Protein Motif Search**
Objective: Identify specific motifs in a sequence.

Tasks:
Search for motifs like phosphorylation sites (e.g., "STQ").


````python
import re

sequence = "MTEYKLVVVGAGGVGKSALTIQLIQNHFVDEYDPTIEDSTQVEDQN"

    ## Compile a pattern for the exact motif 'STQ'
motif = re.compile("STQ")

    ## Find all matches of the motif in the sequence
matches = motif.finditer(sequence)

    ## Print positions of the motif
for match in matches:
    print(f"Motif 'STQ' found at position {match.start()}-{match.end()} in the sequence.")
````
